# 文件处理
## open和with函数

```python
# --- 复制到第 1 个 Cell 中运行 ---
file_name = "my_test.txt"

# 'w' 代表 Write（写入模式）。如果文件不存在，会自动创建；如果存在，会清空原内容！
with open(file_name, 'w', encoding='utf-8') as f:
    f.write("第一行：你好，Jupyter！\n")
    f.write("第二行：这是用来测试的小文件。\n")
    f.write("第三行：🍎 苹果\n")
    f.write("第四行：🍌 香蕉\n")
    f.write("第五行：🍉 西瓜\n")

print("✅ 文件写入成功！你可以去电脑上找找看，是不是多了一个 my_test.txt 文件。")
```

```python
# --- 复制到第 2 个 Cell 中运行 ---
# 'r' 代表 Read（只读模式），这是默认的安全模式。
with open("my_test.txt", 'r', encoding='utf-8') as f:
    # .read() 就像一口气把杯子里的水喝干，把所有文字全装进内存
    content = f.read()

print("✅ 下面是读取到的全部内容：")
print("-" * 20)
print(content)
print("-" * 20)
```

```python
# --- 复制到第 3 个 Cell 中运行 ---
print("✅ 开始逐行读取文件：")
print("-" * 20)

with open("my_test.txt", 'r', encoding='utf-8') as f:
    line_number = 1
    # 这里的 f 就像一个生成器，每次循环只吐出一行数据
    for line in f:
        # .strip() 是为了把每一行末尾自带的换行符去掉，让打印更好看
        clean_line = line.strip() 
        print(f"我拿到了第 {line_number} 行的数据：{clean_line}")
        line_number += 1
```

```python
# --- 复制到第 4 个 Cell 中运行 ---

# 1. 制造一台“流水线机器”（生成器函数）
def mini_lazy_reader(filepath, chunk_size=2):
    with open(filepath, 'r', encoding='utf-8') as f:
        chunk = []
        for line in f:
            chunk.append(line.strip())
            
            # 当水桶装满 2 行时，交出去并暂停
            if len(chunk) == chunk_size:
                yield chunk
                chunk = [] # 清空水桶，为下次做准备
                
        # 兜底：如果最后剩下的不足 2 行，也要交出去
        if len(chunk) > 0:
            yield chunk

# 2. 把机器搬回来（此时还没开始读）
my_generator = mini_lazy_reader("my_test.txt")

# 3. 循环按 next() 按钮，向机器要数据块
chunk_count = 1
for data_chunk in my_generator:
    print(f"📦 收到第 {chunk_count} 个数据块，里面有：{data_chunk}")
    chunk_count += 1
```

# 异常处理（Exception Handling）

```python
# --- 复制到第 1 个 Cell 中 ---
# 【笔记 1：基础防御系统 try...except】

print("--- 测试 A：没有防御 ---")
# 💡 你可以把下面这行前面的 # 删掉试试，程序会直接红字报错，且停止运行
# 10 / 0 

print("\n--- 测试 B：穿上防弹衣 ---")
try:
    print("尝试执行危险的数学动作...")
    result = 10 / 0
    print("（这句话绝对打印不出来，因为上面那行已经出错了）")
except ZeroDivisionError:
    print("🚨 触发警报：成功拦截了“除以零”的错误！")
    print("🔧 备用方案启动，程序没有死机，继续平稳运转。")
```

```python
# --- 复制到第 2 个 Cell 中 ---
# 【笔记 2：抓取详细的诊断报告 (as err)】

try:
    print("尝试打开一个根本不存在的文件...")
    # 'r' 模式下打开不存在的文件必定报错
    with open("我是假文件.txt", "r") as f:
        content = f.read()
        
except FileNotFoundError as err:
    print("🚨 成功拦截 FileNotFoundError (找不到文件)")
    # 打印出 err，就能看到系统底层的原话
    print(f"📄 系统底层的详细诊断报告是：{err}")
```

```python
# --- 复制到第 3 个 Cell 中 ---
# 【笔记 3：主动拉响警报 (raise)】

def withdraw_money(balance, amount):
    print(f"当前余额：{balance}，准备取款：{amount}")
    
    if amount > balance:
        # 遇到不合理的情况，主动按下红色报警按钮！
        # ValueError 专门用来表示“你给的这个值不合法”
        raise ValueError("余额不足，你想抢银行吗！")
        
    return balance - amount

# 💡 尝试调用一下（因为这里没有用 try 保护，所以你会看到真实的红色报错）
# 你可以把 1000 换成 10 试试正常情况下的输出
withdraw_money(50, 1000)
```

```python
# --- 复制到第 4 个 Cell 中 ---
# 【笔记 4：完美配合（工人抛炸弹，老板接炸弹）】
import os

# 1. 工人：只负责干活和检查，发现问题立刻拉响警报 (raise)
def check_and_read(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"工人报告：找不到文件 '{file_path}'")
    return "✅ 工人报告：文件读取成功！"

# 2. 老板：统筹全局，随时准备接炸弹 (try...except)
test_path = "一条错误的路径.txt"

try:
    print("老板：去把文件读出来。")
    result = check_and_read(test_path)
    print(result) # 如果顺利，就会打印工人的成功报告
    
except FileNotFoundError as e:
    print(f"老板：稳稳接住炸弹。收到的工人警报是 -> {e}")
    print("老板：启动 B 计划，提示用户重新检查路径。")
```

```python
# 【笔记 5：多重防御与兜底机制】

try:
    print("老板：员工，去把文件里的数字读出来，然后用 100 去除以它！")
    
    # 动作 1：打开文件（可能会触发 FileNotFoundError）
    with open("data.txt", "r") as f:
        content = f.read()
        
    # 动作 2：把文字变成整数（可能会触发 ValueError，比如里面写的是"苹果"）
    number = int(content)
    
    # 动作 3：做除法计算（可能会触发 ZeroDivisionError，比如里面写的是 0）
    result = 100 / number
    
    print(f"计算成功！结果是 {result}")

# 🏥 开始分诊：
except FileNotFoundError:
    print("🚑 寻人科：文件根本不存在，请检查路径！")

except ValueError:
    print("🚑 内分泌科：文件里写的不是数字，无法转换，请检查数据！")

except ZeroDivisionError:
    print("🚑 数学科：文件里的数字是 0，不能作为除数！")

# 👑 终极绝招：兜底机制
except Exception as e:
    print(f"🚨 院长亲自出马：发生了一个极其罕见的未知错误，具体原因是：{e}")
```
